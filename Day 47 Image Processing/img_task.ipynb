```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("fruit.png")
```

```python
def show(img):
    
    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    else:                     # grayscale image
        plt.imshow(img, cmap='gray')

    plt.axis("off")
    plt.show()
```

```python
show(img)
```

```python
resize = cv2.resize(img, (300, 300))

print("Original shape:", img.shape)
print("Resized shape:", resize.shape)

show(resize)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

show(gray)
```

```python
blur = cv2.GaussianBlur(gray, (5, 5), 0)

show(blur)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 100, 200)

show(edges)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, 100, 200)
dilation = cv2.dilate(edges, kernel=(3,3), iterations=1)

show(dilation)
```

```python
print(type(img))
print(type(gray))
print(type(edges))
print(type(dilation))
```

```python
shows = [img, gray, edges, dilation]

for i in shows:
    show(i)
```
