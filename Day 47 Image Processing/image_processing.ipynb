```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("fruit.png")
```

```python
plt_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow("Image", img)
cv2.waitKey(0)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("Image", img_gray)
cv2.waitKey(0)
```

```python
plt_img_gray = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2RGB)
```

```python
plt.imshow(plt_img_gray, cmap="gray")
plt.axis("off")
plt.show()
```

```python
blue = img.copy()
green = img.copy()
red = img.copy()
```

```python
blue[:,:,0] = 0

cv2.imshow("Image", blue)
cv2.waitKey(0)
```

```python
green[:,:,1] = 0

cv2.imshow("Image", green)
cv2.waitKey(0)
```

```python
red[:,:,2] = 0

cv2.imshow("Image", red)
cv2.waitKey(0)
```

```python
img_blue = blue
img_green = green
img_red = red

new_img = np.hstack((img_blue, img_green, img_red))
```

```python
cv2.imshow("Image", new_img)
cv2.waitKey(0)
```

```python
plt_new_img = cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(plt_new_img, cmap="gray")
plt.axis("off")
plt.show()
```
