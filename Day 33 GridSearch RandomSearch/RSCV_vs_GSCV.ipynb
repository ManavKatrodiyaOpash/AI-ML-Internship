# RandomSearchCV vs GridSearchCV

```python
from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report, r2_score
```

# Classification

```python
x_clf,y_clf = load_breast_cancer(return_X_y=True)
```

```python
x_clf
```

```python
y_clf
```

```python
x_clf_train, x_clf_test, y_clf_train, y_clf_test = train_test_split(x_clf, y_clf, test_size=0.2, random_state=42)
```

```python
param_grid_clf = {
    "n_estimators": [50, 100, 200],
    "max_depth": [5, 10, 15],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 5, 10],
    "criterion": ["gini", "entropy"],
}
```

```python
rf_clf = RandomForestClassifier()
```

## GridSearchCV

```python
gs_clf = GridSearchCV(
    estimator=rf_clf,
    param_grid=param_grid_clf,
    scoring="accuracy",
    cv=5,
    return_train_score=True,
)
```

```python
gs_clf.fit(x_clf_train, y_clf_train)
```

```python
gs_clf.best_params_
```

```python
rf_gs_clf = RandomForestClassifier(
    **gs_clf.best_params_,
    random_state=42
)
```

```python
rf_gs_clf.fit(x_clf_train, y_clf_train)
```

```python
y_clf_gs_pred = rf_gs_clf.predict(x_clf_test)
```

```python
accuracy_score(y_clf_test, y_clf_gs_pred)
```

```python
print(classification_report(y_clf_test, y_clf_gs_pred))
```

## RandomSearchCV

```python
rs_clf = RandomizedSearchCV(
    estimator=rf_clf,
    param_distributions=param_grid_clf,
    scoring="accuracy",
    cv=5,
    return_train_score=True,
)
```

```python
rs_clf.fit(x_clf_train, y_clf_train)
```

```python
rs_clf.best_params_
```

```python
rf_rs_clf = RandomForestClassifier(
    **rs_clf.best_params_,
    random_state=42
)
```

```python
rf_rs_clf.fit(x_clf_train, y_clf_train)
```

```python
y_clf_rs_pred = rf_rs_clf.predict(x_clf_test)
```

```python
accuracy_score(y_clf_test, y_clf_rs_pred)
```

```python
print(classification_report(y_clf_test, y_clf_rs_pred))
```

# Regression

```python
x_reg,y_reg = load_diabetes(return_X_y=True)
```

```python
x_reg
```

```python
y_reg
```

```python
x_reg_train, x_reg_test, y_reg_train, y_reg_test = train_test_split(x_reg, y_reg, test_size=0.2, random_state=42)
```

```python
param_grid_reg = {
    "n_estimators": [50, 100, 200],
    "max_depth": [5, 10, 15],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 5, 10],
    "criterion" : ["squared_error", "friedman_mse"]
}
```

```python
rf_reg = RandomForestRegressor()
```

## GridSearchCV

```python
gs_reg = GridSearchCV(
    estimator=rf_reg,
    param_grid=param_grid_reg,
    cv=5,
    return_train_score=True,
)
```

```python
gs_reg.fit(x_reg_train, y_reg_train)
```

```python
gs_reg.best_params_
```

```python
rf_gs_reg = RandomForestRegressor(
    **gs_reg.best_params_,
    random_state=42
)
```

```python
rf_gs_reg.fit(x_reg_train, y_reg_train)
```

```python
y_reg_gs_pred = rf_gs_reg.predict(x_reg_test)
```

```python
r2_score(y_reg_test, y_reg_gs_pred)
```

## RandomSearchCV

```python
rs_reg = RandomizedSearchCV(
    estimator=rf_reg,
    param_distributions=param_grid_reg,
    cv=5,
    return_train_score=True,
)
```

```python
rs_reg.fit(x_reg_train, y_reg_train)
```

```python
rs_reg.best_params_
```

```python
rf_rs_reg = RandomForestRegressor(
    **rs_reg.best_params_,
    random_state=42
)
```

```python
rf_rs_reg.fit(x_reg_train, y_reg_train)
```

```python
y_reg_rs_pred = rf_rs_reg.predict(x_reg_test)
```

```python
r2_score(y_reg_test, y_reg_rs_pred)
```
