# Random Forest Theory on Regression

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_regression
from sklearn.tree import DecisionTreeRegressor, plot_tree
```

```python
x, y = make_regression(
    n_samples=100,
    n_features=5,
    noise=0.1,
    random_state=42
)
```

```python
df = pd.DataFrame(x, columns=[f'col{i + 1}' for i in range(x.shape[1])])
df['target'] = y
```

```python
df
```

```python
model1 = DecisionTreeRegressor()
model2 = DecisionTreeRegressor()
model3 = DecisionTreeRegressor()
```

```python
model1.fit(x[:70], y[:70])
model2.fit(x[30:100], y[30:100])
model3.fit(x[50:], y[50:])
```

```python
y_pred1 = model1.predict(np.array([0.975120, -0.677162, -0.012247, -0.897254, 0.075805]).reshape(1, -1))
y_pred1
```

```python
y_pred2 = model2.predict(np.array([0.975120, -0.677162, -0.012247, -0.897254, 0.075805]).reshape(1, -1))
y_pred2
```

```python
y_pred3 = model3.predict(np.array([0.975120, -0.677162, -0.012247, -0.897254, 0.075805]).reshape(1, -1))
y_pred3
```

```python
y_pred = (y_pred1 + y_pred2 + y_pred3) / 3
y_pred
```

```python
plot_tree(model1)
```

```python
plot_tree(model2)
```
