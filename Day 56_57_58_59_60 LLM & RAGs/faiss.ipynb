```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
list = ["I love playing Volleyball", "I love Watching Movies and Series", "Coding Python is very fun", "Dhosa is my favourite food", "Time Travel Movies are my favourite"]
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(list)
```

```python
embeddings
```

```python
embeddings.shape
```

```python
dimentions = embeddings.shape[1]
```

```python
index = faiss.IndexFlatL2(dimentions)
```

```python
index.add(np.array(embeddings))
```

```python
query = "I like watching films"
query_embedding = model.encode([query])
```

```python
k = 1
distances, indices = index.search(np.array(query_embedding), k)
```

```python
print("Query:", query)
print("Most similar sentence:", list[indices[0][0]])
```

```python
k = 3
distances, indices = index.search(np.array(query_embedding), k)

print("Query:", query)

for i in range(k):
    print(f"Result {i+1}:", list[indices[0][i]])
```
