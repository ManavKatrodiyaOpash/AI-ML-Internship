```python
from transformers import pipeline
```

```python
text_generator = pipeline('text-generation', model='gpt2')
```

```python
prompt = "I am Happy to learn about LLMs because"
```

```python
result = text_generator(prompt, max_length=50, num_return_sequences=1)

print(result[0]['generated_text'])
```

```python
sentiment_analyzer = pipeline('sentiment-analysis')
```

```python
pr = "I Love this Movie"
```

```python
result = sentiment_analyzer(pr)

print(result[0]["label"], round(result[0]["score"]))
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

def translator(text):
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model.generate(**inputs)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
translator("I am learning about LLMs")
```

```python
translator("I am Happy")
```

```python
translator("Now i am not")
```
