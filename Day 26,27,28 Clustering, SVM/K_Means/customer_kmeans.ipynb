```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
x = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
plt.scatter(
    x['Annual Income (k$)'],
    x['Spending Score (1-100)'],
)
plt.title("Before Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.tight_layout()
plt.show()
```

```python
km = KMeans(n_clusters=5)
km.fit(x)
label = km.labels_
```

```python
plt.scatter(
    x['Annual Income (k$)'],
    x['Spending Score (1-100)'],
    c=label,
)
plt.title("After Clustering")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.tight_layout()
plt.show()
```

```python
x_age_income = df[['Age','Annual Income (k$)']]
```

```python
plt.scatter(
    x_age_income['Age'],
    x_age_income['Annual Income (k$)'],
)
plt.title("Age vs Annual Income (Before Clustering)")
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.tight_layout()
plt.show()
```

```python
km_age_income = KMeans(n_clusters=5)
km_age_income.fit(x_age_income)
age_income_label = km_age_income.labels_
```

```python
plt.scatter(
    x_age_income['Age'],
    x_age_income['Annual Income (k$)'],
    c=age_income_label,
)
plt.title("Age vs Annual Income (After Clustering)")
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.tight_layout()
plt.show()
```

```python
x_age_spending = df[['Age','Spending Score (1-100)']]
```

```python
plt.scatter(
    x_age_spending['Age'],
    x_age_spending['Spending Score (1-100)'],
)
plt.title("Age vs Spending Score (1-100) (Before Clustering)")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.tight_layout()
plt.show()
```

```python
plt.scatter(
    x_age_spending['Age'],
    x_age_spending['Spending Score (1-100)'],
    c=age_spending_label,
)
plt.title("Age vs Spending Score (1-100) (After Clustering)")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.tight_layout()
plt.show()
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit(x)
    wcss.append(km.inertia_)
```
