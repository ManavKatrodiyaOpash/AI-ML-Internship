```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chaat of delhi")

for token in doc.sents:
    print(token)
```

```python
nlp = spacy.blank("en")

doc = nlp("Mr. Sharma visits mumbai and loved the vadapav there The cost was 2$")

for token in doc:
    print(token)
```

```python
nlp_hin = spacy.blank("hi")

doc = nlp_hin("कैसे हो मेरे दोस्त सब खुशल मङल ?")

for token in doc:
    print(token)
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme", [
    {ORTH : "gim"},
    {ORTH : "me"}
])

doc = nlp("gimme double cheese healthy pizza")
tokens = [token for token in doc]
tokens
```
