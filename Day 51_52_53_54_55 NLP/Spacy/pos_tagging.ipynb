```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Elon flew to mars yesterday. He carried biryani masala with him")

for token in doc:
    print(f"{token} | {token.pos_} | {spacy.explain(token.pos_)}")
print("\n")
for token in doc:
    print(f"{token} | {token.tag_} | {spacy.explain(token.tag_)}")
```
