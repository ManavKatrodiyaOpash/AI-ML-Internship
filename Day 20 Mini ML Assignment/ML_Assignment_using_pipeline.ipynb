```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df = df.drop(columns=['EmployeeID'])
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
bin_cols = ['Gender','Overtime']
ord_cols = ['Education','StressLevel']
nom_cols = ['Department','JobRole']
```

```python
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
ord_cols = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder())
])
```

```python
nom_cols = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ("encoder", OneHotEncoder(categories=[
        ["Bachelor", "Master", "PhD"],
        ["Low", "Medium", "High"]
    ]))
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, num_cols),
    ('ord', ord_cols, ord_cols),
    ('nom', nom_cols, nom_cols)
])
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])
```

```python
x = df.drop(columns=["Attrition"],axis=1)
y = df['Attrition'] 
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,train_size=0.2, random_state=42
)
```
