```python
from random import shuffle

import pandas as pd
from matplotlib import pyplot as plt

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold, StratifiedKFold, LeaveOneOut, cross_val_score, cross_validate
```

```python
x, y = load_iris(return_X_y=True)
```

```python
model = LogisticRegression(max_iter=200)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
stffold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
loo = LeaveOneOut()
```

```python
models = {
    "K-Fold": kfold,
    "Stratified K-Fold": stffold,
    "Leave-One-Out": loo
}
```

```python
result = []

for name, mod in models.items():
    score = cross_val_score(model, x, y, cv=mod)
    result.append({
        "model": name,
        "cross-val-score": score.mean(),
    })
df = pd.DataFrame(result)
```

```python
df
```

```python
cv_result = []

for name, mod in models.items():
    score = cross_validate(
    model,
    x,
    y,
    cv=mod,
    scoring={
        "accuracy": "accuracy",
    }
)
    cv_result.append({
        "model": name,
        "Mean CV Score": score["test_accuracy"].mean(),
    })
df = pd.DataFrame(cv_result)
```
