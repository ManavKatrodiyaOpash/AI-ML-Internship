# Importing Libraries

```python
from sklearn.datasets import load_iris, make_regression
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import cross_val_score

from sklearn.neighbors import KNeighborsClassifier
```

# Cross Validation Classification

```python
x_clf,y_clf = load_iris(return_X_y=True)
```

## Logistic Regression

```python
log_model = LogisticRegression(max_iter=200)
```

```python
log_score = cross_val_score(log_model, x_clf, y_clf, cv=10)
```

```python
print(log_score)
print(log_score.mean())
```

## K Nearest Neighbors Classifier

```python
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn_score = cross_val_score(knn, x_clf, y_clf, cv=10)
    print(i, knn_score)
```

# Cross Validation Regression

```python
x_reg, y_reg = make_regression(n_samples=1000, n_features=5, noise=0.1, random_state=0)
```

## Linear Regression

```python
lin_model = LinearRegression()
```

```python
lin_score = cross_val_score(lin_model, x_reg, y_reg, cv=10)
```

```python
print(lin_score)
print(lin_score.mean())
```
