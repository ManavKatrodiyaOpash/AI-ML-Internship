```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import silhouette_score

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN,KMeans
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
db = DBSCAN(eps=0.3, min_samples=5)
labels = db.fit_predict(x_scaled)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels)
plt.tight_layout()
plt.show()
```

```python
km = KMeans(n_clusters=5, random_state=1)
labels = km.fit_predict(x_scaled)
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels)
plt.tight_layout()
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors

neighbors = NearestNeighbors(n_neighbors=5) # n_neighbors should match your min_samples
neighbors_fit = neighbors.fit(x_scaled)
distances, indices = neighbors_fit.kneighbors(x_scaled)

distances = np.sort(distances[:, 4], axis=0)
plt.plot(distances)
plt.title("K-Distance Graph")
plt.ylabel("Epsilon Value")
plt.show()
```

```python
eps_val = np.round(np.arange(0.1,0.6,0.05),3)
min_sample_val = [3,4,5,6,7]

result = []

for eps in eps_val:
    for ms in min_sample_val:
        db = DBSCAN(eps=eps, min_samples=ms)
        labels = db.fit_predict(x_scaled)

        unique_labels = set(labels)
        n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)
        n_noise = np.sum(labels == -1)
        score = silhouette_score(x_scaled,labels)
        print(f"eps={eps} | min_sample = {ms} | clusters={n_clusters} | noise points={n_noise} | silhouette_score={score}")

        plt.scatter(x_scaled[:,0],x_scaled[:,1],c=labels)
        plt.title(f"Epsilon :- {eps} and Min Samples :- {ms}")
        plt.tight_layout()
        plt.show()
        print("-"*100)

        result.append({
            "eps": eps,
            "min_sample" : ms,
            "clusters": n_clusters,
            "noise_points": n_noise,
            "silhouette_score": score
        })
df_score = pd.DataFrame(result)
```

```python
df_score
```

```python
df_score[
    (df_score['clusters'] > 2) &
    (df_score['min_sample'] >= 5) &
    (df_score['noise_points'] < 100) &
    (df_score['silhouette_score'] > 0)

].sort_values(by='silhouette_score',ascending=False)
```

**By Observation epsilon = 0.4, min_sample = 5, clusters=4, noise_points=15 is best with the silhouette_score = 0.413258

```python
db_final = DBSCAN(eps=0.4,min_samples=5)
labels = db_final.fit_predict(x_scaled)

unique_labels = set(labels)

for label in unique_labels:
    if label == -1:
        plt.scatter(
            x_scaled[labels == label,0],
            x_scaled[labels == label,1],
            c='black',
            label = 'Noise',
        )
    else:
        plt.scatter(
            x_scaled[labels == label,0],
            x_scaled[labels == label,1],
            label=f"Cluster {label+1}",
            s=30
        )

plt.title("Best Plot according to received value")
plt.xlabel("Annual Income (Scaled)")
plt.ylabel("Spending Score (Scaled)")
plt.tight_layout()
plt.legend()
plt.show()
```
